## Seleções de Frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
#Selecionar somente os imóveis classificados como tipo 'Apartamento'
selecao = dados['Tipo'] == 'Apartamento'
selecao
```

```python
n1 = dados[selecao]
```

```python
d1 = n1.shape[0]
```

```python
n1.head(10)
```

```python
# Selecione os imóveis classificados como tipo 'Casa', 'Casa de Condominio' e 'Casa de Vila'
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
selecao
```

```python
n2 = dados[selecao]
```

```python
d2 = n2.shape[0]
```

```python
n2.head(10)
```

```python
#Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites
```

```python
# 60 <= Area <=100
selecao = (dados['Area'] >=60) & (dados['Area'] <=100)
selecao
```

```python
n3 = dados[selecao]
```

```python
d3 = n3.shape[0]
```

```python
n3.head(10)
```

```python
# Selecione os imóveis que tenha pelo menos 4 quartos e aluguel menor que R$ 2.000,00
```

```python
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecao]
```

```python
n4
```

```python
n4.head(10)
```

```python
d4 = n4.shape[0]
```

```python
print("Nº de imóveis classificados como tipo 'Apartamento'-> {}" .format(d1))
print("Nº de imóveis classificados como tipo 'Casa', 'Casa de Condominio' e 'Casa de Vila'-> {}" .format(d2))
print("Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites-> {}" .format(d3))
print("Nº de imóveis que tenha pelo menos 4 quartos e aluguel menor que R$ 2.000,00-> {}" .format(d4))
```
